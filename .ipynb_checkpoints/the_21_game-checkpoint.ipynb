```python
import random as rd

number = 0
current_player = 'human' if rd.randint(0,1)==0 else 'computer'
while number<=21:
    if current_player == 'human':
        print('Number: ',number)
        player_choice=''
        cond=False
        while cond == False:
            player_choice = input('Human choose: ')
            if player_choice in ['1','2','3']:
                cond = True
                break
            else:
                cond = False
        if cond:
            number += int(player_choice)
        if number >=21:
            print('Human lose')
            play_again = input('\nDo you want play again? y/n? ')
            if play_again == 'y':
                number = 0
            else:
                break
        else:
            current_player= 'computer'
    if current_player == 'computer':
        print('Number: ',number)
        com_choice = rd.randint(1,3)
        print('Computer choose: ',com_choice)
        number += com_choice
        if number >=21:
            print('Computer lose')
            play_again = input('\nDo you want play again? y/n? ')
            if play_again == 'y':
                number = 0
            else:
                break
        else:
            current_player= 'human'
      
```
