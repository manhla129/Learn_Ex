```python
class sanpham:
    def __init__(self, ms , name):
        self.ms = ms
        self.name = name
        
    def infor(self):
        print(f'{self.ms}\t\t{self.name}')
        
class dssanpham:
    def __init__(self, _list):
        self.list = _list
        
    def initialize(self):
        sp1 = sanpham('101','Milk')
        sp2 = sanpham('102','Sugar')
        sp3 = sanpham('103','Candy')
        self.list = [sp1, sp2, sp3]
        
    def output(self):
        for i in self.list:
            i.infor()
            
    def add(self, sp):
        self.list.append(sp)
        
    def delete(self, ms):
        for i in self.list:
            if i.ms == ms:
                self.list.remote(i)
                break
    
    def rename(self, ms, sp):
        for i in self.list:
            if i.ms == ms:
                i.name = sp.name
                
class Quanlysanpham:
    def Menu(self):
        print('----------Quản lý sản phẩm-----------')
        print('--1. Tải danh sách                 --')
        print('--2. Hiển thị danh sách            --')
        print('--3. Thêm sản phẩm                 --')
        print('--4. Xóa sản phẩm                  --')
        print('--5. Sửa tên sản phẩm              --')
        print('--0.Thoát                          --')
        print('-------------------------------------')
        
    list_sp = None
    def Run(self):
        while True:
            print()
            self.Menu()
            action = input('Nhập lựa chọn:')
            if action == '1':
                self.list_sp = dssanpham(None)
                self.list_sp.initialize()
                print('Tải thành công')
                
            elif action == '2':
                if self.list_sp is None:
                    print('Chưa tải danh sách')
                else:
                    self.list_sp.output()
            elif action == '3':
                ms = input('Nhập mã sản phẩm: ')
                name = input('Nhập tên: ')
                sp = sanpham(ms, name)
                self.list_sp.add(sp)
                print('Thêm thành công')
            elif action == '4':
                ms = input('Nhập mã sản phẩm: ')
                self.list_sp.delete(ms)
                print('Xóa thành công')
            elif action == '5':    
                ms = input('Nhập mã sản phẩm: ')
                name = input('Nhập tên: ')
                sp = sanpham(ms, name)
                self.list_sp.rename(ms, sp)
                print('Sửa thành công')
            elif action == '0':
                break
sm = Quanlysanpham()
sm.Run()
                
```
