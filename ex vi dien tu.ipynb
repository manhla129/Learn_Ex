```python
def find_object(my_list, find_ob):
    index = -1
    for i in range(len(my_list)):
        if my_list[i] == find_ob:
            index = i
    return index

def add_object(my_list, new_ob):
    if find_object(my_list, new_ob) >= 0:
        print('Đã có mục chi tiêu này')
    else:
        my_list.append(new_ob)
        
def del_object(my_list, del_ob):
    if find_object(my_list, new_ob) < 0:
        print('Không có có mục chi tiêu này')
    else:
        my_list.remove(del_ob)
        
def menu():
    print('------MENU------')
    print('-1. Find -------')
    print('-2. Add --------')
    print('-3. Del --------')
    print('-0. Esc --------')
    print('----------------')
    
def main():
    my_list = ['edu', 'fashion', 'food', 'traffic']
    while True:
        menu()
        action = input('Lựa chọn 1 mục dưới đây: ')
        if action == '1':
            find_ob = input('Nhập mục tìm kiếm')
            print(find_object(my_list, find_ob))
        elif action == '2':
            new_ob = input('Nhập mục muốn thêm')
            add_object(my_list, new_ob)
        elif action == '3':
            del_ob = input('Nhập mục muốn xóa')
            del_object(my_list, del_ob)
        else:
            break
main()
```
