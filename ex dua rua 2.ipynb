```python
import turtle as tl
import random
import timeit

screen = tl.Screen()
screen.setup(height=500, width=600)
pen = tl.Turtle(visiable=False)
pen.penup()
pen.speed(30)
pen.goto(-250, 200)

for i in range(21):
    pen.write(i)
    pen.forward(25)
    
x = -250
pen.goto(-250, 200)
pen.right(90)
for i in range(21):
    for j in range(10):
        pen.pendown()
        pen.forward(20)
        pen.penup()
        pen.forward(10)
    pen.penup()
    pen.forward(5)
    pen.write(i)
    pen.goto(x + (i+1)*25, 200)
    
all_turtle = []
y_pos = [160, 100, 40, -20]
colors = ['red','green','blue','black']
for i in range(4):
    turtles = tl.Turtle(shape='turtle')
    turtles.penup()
    turtles.goto(x=-250, y=y_pos[i])
    turtles.color(colors[i])
    all_turtle.append(turtles)
    
def find_id(list_item, find_item):
    index=-1
    for i in range(len(list_item)):
        if list_item[i] = find_item:
            index = i
            break
    return index
    
def random_walk(turtles):
    global run
    for i in turtles:
        i.forward(random.randint(1,5))
        if i.xcor() > 250:
            run = False
            print(f'{colors[find_id(turtles, i)]} turtle is Winner!')
            stop = timeit.default_timer()
            print('Time run: ', stop-start)
            
run = True
global start
start = timeit.default_timer()
while run:
    random_walk(all_turtle)
    
screen.exitonclick()
```
